```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('compas_results_dmc.csv', sep=';')
#df = pd.read_csv('results/compas_results_odm.csv', sep=';')

df *= 100
df['epsilon'] /=100
```

```python
#FAIRNESS
plt.plot(df['epsilon'],df['FAIR_RUXRF'], color='steelblue', label='FairRUXRF', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_RUXADA'], color='lightskyblue', label='FairRUXADA', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_RUXGB'], color='navy',label='FairRUXGB', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_RUG'], color='dodgerblue', label='FairRUG', linewidth=0.9)

plt.plot(df['epsilon'],df['FAIR_RF'], color='indianred', linestyle='dashed', label='RF', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_ADA'], color='firebrick', linestyle='dashed', label='ADA', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_GB'], color='red', linestyle='dashed',  label='GB', linewidth=0.9)
plt.plot(df['epsilon'],df['FAIR_DT'], color='rosybrown', linestyle='dashed', label='DT', linewidth=0.9)

plt.xlabel('Allowed level of unfairness \u03B5')
plt.ylabel('Fairness (%)')

plt.legend(loc='upper right', ncol=2, fontsize='medium')
plt.title('Disparate Mistreatment per Class')
#plt.title('Overall Disparate Mistreatment')
plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams['font.sans-serif'] = "Georgia"
plt.show()
```

```python
#ACCURACY
plt.plot(df['epsilon'],df['ACC_RUXRF'], color='steelblue', label='FairRUXRF', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_RUXADA'], color='lightskyblue', label='FairRUXADA', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_RUXGB'], color='navy',label='FairRUXGB', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_RUG'], color='dodgerblue', label='FairRUG', linewidth=0.9)

plt.plot(df['epsilon'],df['ACC_RF'], color='indianred', linestyle='dashed', label='RF', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_ADA'], color='firebrick', linestyle='dashed', label='ADA', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_GB'], color='red', linestyle='dashed',  label='GB', linewidth=0.9)
plt.plot(df['epsilon'],df['ACC_DT'], color='rosybrown', linestyle='dashed', label='DT', linewidth=0.9)

plt.xlabel('Allowed level of unfairness \u03B5')
plt.ylabel('Accuracy (%)')
plt.legend(loc='lower right', ncol=2, fontsize='medium')
plt.title('Disparate Mistreatment per Class')
#plt.title('Overall Disparate Mistreatment')
plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams['font.sans-serif'] = "Georgia"
plt.show()
```
